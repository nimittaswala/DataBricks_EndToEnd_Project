```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
init_flag = int(dbutils.widgets.get("init_flag"))
```

### Data Reading from Source

```python
df = spark.sql("select * from databricks_cata.silver.customers_silver")
```

```python
df.display()
```


### Removing Duplicates

```python
df = df.dropDuplicates(subset=['customer_id'])
df.limit(10).display()
```

### Dividing New vd Old Records

```python
if init_flag == 0:

    df_old = spark.sql('''select DimCustomerKey, customer_id, create_date, update_date from databricks_cata.gold.DimCustomers''')
    
else:

    df_old = spark.sql(''' select 0 DimCustomerKey, 0 customer_id, 0 create_date, 0 update_date from databricks_cata.silver.customers_silver where 1=0''')
```

```python
df_old.limit(10).display()
```

### Renaming column of df_old

```python
df_old = df_old.withColumnRenamed('DimCustomerKey','old_DimCustomerKey')\
                              .withColumnRenamed('customer_id','old_customer_id')\
                              .withColumnRenamed('create_date','old_create_date')\
                              .withColumnRenamed('update_date','old_update_date')
```

### Applying Join with old records

```python

 df_join = df.join(df_old, df['customer_id'] == df_old['old_customer_id'], 'left')
```

```python
df_join.display()
```

### Separating New vs Old Records

```python
df_new = df_join.filter(df_join['old_DimCustomerKey'].isNull())
```

```python
df_old = df_join.filter(df_join['old_DimCustomerKey'].isNotNull())
```

### preparing df_old

```python
# Dropping all the columns which are not required

df_old = df_old.drop('old_customer_id','old_update_date')

# renaming from old_DimCustomerKey to DimCustomerKey

df_old = df_old.withColumnRenamed('old_DimCustomerKey','DimCustomerKey')

# Renaming "old_create_date" to "create_date"

df_old = df_old.withColumnRenamed('old_create_date','create_date')
df_old = df_old.withColumn('create_date',to_timestamp(col('create_date')))

# Recreating  the "update" column with current timestamp

df_old = df_old.withColumn('update_date',current_timestamp())
```

```python
df_old.display()
```

### preparing df_new

```python
# Dropping all the columns which are not required

df_new = df_new.drop('old_customer_id','old_update_date','old_create_date','old_DimCustomerKey')

# Recreating  the "update_date","current_date" column with current timestamp

df_new = df_new.withColumn('update_date',current_timestamp())
df_new = df_new.withColumn('create_date',current_timestamp())

```

```python
df_new.display()
```

### Surrogate key -from 1

```python
df_new = df_new.withColumn('DimCustomerKey',monotonically_increasing_id() + lit(1))
```

```python
df_new.display()
```

### Adding Max Surrogate Key

```python
if init_flag == 1:
    max_surroage_key = 0
else:
    df_maxsur = spark.sql("select max(DimCustomerKey) as max_surroagate_key from databricks_cata.gold.DimCustomers")

    # converting df_maxsur to max_surrogate_key variable
    max_surroage_key = df_maxsur.collect()[0]['max_surroagate_key']
```

```python
df_new = df_new.withColumn('DimCustomerKey',lit(max_surroage_key) + col('DimCustomerKey'))
```

### Union of df_old and df_new

```python
df_final = df_new.unionByName(df_old)
```

```python
df_old.display()
```

```python
display(df_final)
```

### SCD Type - 1

```python
from delta.tables import DeltaTable
```

```python
if (spark.catalog.tableExists('databricks_cata.gold.DimCustomers')):

    dlt_obj = DeltaTable.forPath(spark,"abfss://gold@databricksetenimit.dfs.core.windows.net/DimCustomers")

    dlt_obj.alias('trg').merge(df_final.alias('src'),'trg.DimCustomerKey = src.DimCustomerKey')\
        .whenMatchedUpdateAll()\
        .whenNotMatchedInsertAll()\
        .execute()
    
else:
    df_final.write.mode("overwrite")\
    .format("delta")\
    .option("path","abfss://gold@databricksetenimit.dfs.core.windows.net/DimCustomers")\
    .saveAsTable("databricks_cata.gold.DimCustomers")
    
```
