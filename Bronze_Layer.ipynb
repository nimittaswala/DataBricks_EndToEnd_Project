### Dynamic Capabilities

```python
dbutils.widgets.text("file_name","")
```

```python
P_file_name = dbutils.widgets.get("file_name")
```

### Data Reading

```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "parquet")\
    .option("cloudFiles.schemaLocation", f"abfss://bronze@databricksetenimit.dfs.core.windows.net/checkpoint_{P_file_name}")\
    .load(f"abfss://source@databricksetenimit.dfs.core.windows.net/{P_file_name}")
    
```

### Data Writing

```python
df.writeStream.format("Parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@databricksetenimit.dfs.core.windows.net/checkpoint_{P_file_name}")\
    .option("path", f"abfss://bronze@databricksetenimit.dfs.core.windows.net/{P_file_name}")\
    .trigger(once=True)\
    .start()
```
