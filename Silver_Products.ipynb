```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

### Data Reading

```python
df = spark.read.format("parquet").load("abfss://bronze@databricksetenimit.dfs.core.windows.net/products")
```

```python
display(df)
```

```python
df= df.drop("_rescued_data")
display(df)
```

```python
df.createOrReplaceTempView("products")
```

### Functions

```python
%sql
create or replace  function databricks_cata.bronze.discount_func(p_price double)
returns double
language sql
return p_price * 0.90
```

```python
%sql
select product_id,price, databricks_cata.bronze.discount_func(price)  as discount_price
from products
```

```python
from pyspark.sql.functions import expr
df = df.withColumn("discounted_price", expr("databricks_cata.bronze.discount_func(price)"))
df.display()
```
